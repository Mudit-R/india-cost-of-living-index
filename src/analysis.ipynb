# Cost of Living Index Analysis - 50 Indian Cities

This notebook provides an interactive analysis of cost of living across Indian cities.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from data_loader import DataLoader
from cost_calculator import CostIndexCalculator
from visualizer import Visualizer

sns.set_style('whitegrid')
%matplotlib inline
```

## 1. Load Data

```python
loader = DataLoader()

fuel_df = loader.load_fuel_prices()
uber_df = loader.load_uber_prices()
doctor_df = loader.load_doctor_fees()
grocery_df = loader.load_blinkit_data()
housing_df = loader.load_housing_data()

print(f"Loaded data for {len(loader.cities)} unique cities")
```

## 2. Calculate Cost Index

```python
calculator = CostIndexCalculator()
merged_df = calculator.merge_all_data(fuel_df, uber_df, doctor_df, grocery_df, housing_df)
result_df = calculator.calculate_index(merged_df, base_city='Delhi')

result_df.head(10)
```

## 3. Summary Statistics

```python
stats = calculator.get_summary_stats(result_df)
for key, value in stats.items():
    print(f"{key}: {value}")
```

## 4. Visualizations

```python
# Top 15 cities
top_15 = result_df.head(15)
plt.figure(figsize=(12, 6))
plt.barh(top_15['City'], top_15['cost_of_living_index'], color='coral')
plt.xlabel('Cost of Living Index')
plt.title('Top 15 Most Expensive Cities')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Component comparison
components = ['housing_index', 'grocery_index', 'transport_index', 'healthcare_index', 'fuel_index']
top_10 = result_df.head(10)

fig, ax = plt.subplots(figsize=(14, 6))
top_10[components].plot(kind='bar', ax=ax, width=0.8)
ax.set_xticklabels(top_10['City'], rotation=45, ha='right')
ax.set_ylabel('Index Value')
ax.set_title('Cost Component Breakdown - Top 10 Cities')
ax.legend(['Housing', 'Grocery', 'Transport', 'Healthcare', 'Fuel'])
ax.axhline(y=100, color='black', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

## 5. Export Results

```python
result_df.to_csv('cost_index_results.csv', index=False)
print("Results saved to cost_index_results.csv")
```
